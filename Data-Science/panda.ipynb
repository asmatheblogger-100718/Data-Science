```python
import pandas as pd

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Check missing values BEFORE any operation
print("Missing values BEFORE handling:")
print(df.isna().sum())

# Fill missing Age values with 0
df["Age"] = df["Age"].fillna(0)

print("\nMissing Age values AFTER fillna:")
print(df["Age"].isna().sum())

# Drop rows where ANY column still has missing values
df = df.dropna()

print("\nMissing values AFTER dropna:")
print(df.isna().sum())

# Dataset shape after cleaning
print("\nDataset shape after fillna + dropna:")
print(df.shape)

# Preview cleaned data
print("\nFirst 5 rows of cleaned dataset:")
print(df.head())
```

```python
#imputation technique
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Separate known and missing Age values
age = df["Age"].dropna()
missing_age = df["Age"].isna()

# Print known Age values (sample)
print("Known Age values (first 10):")
print(age.head(10))

print("\nNumber of missing Age values:")
print(missing_age.sum())

# Randomly sample existing values
random_ages = np.random.choice(age,size=missing_age.sum(),replace=True)

print("\nRandomly generated Age values for imputation (first 10):")
print(random_ages[:10])



# Impute missing values
df.loc[missing_age, "Age"] = random_ages
# Verify imputation
print("\nMissing Age values AFTER imputation:")
print(df["Age"].isna().sum())

print("\nUpdated Age column (first 10 rows):")
print(df["Age"].head(10))
```

```python
#4.	Fix data types (convert to numeric, datetime, category).
import pandas as pd

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

print("=== Original Data Types ===")
print(df.dtypes)

#Convert to Numeric
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")#errors=coerce:force invalid data into NaN
df["Fare"] = pd.to_numeric(df["Fare"], errors="coerce")

# Convert to Datetime
# (Example: create a dummy date column)
df["Sample_Date"] = pd.to_datetime("2024-01-01")

#  Convert to Category
df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
df["Pclass"] = df["Pclass"].astype("category")

# Display Updated Data Types
print("\n=== Updated Data Types ===")
print(df.dtypes)

# Show Sample Data
print("\n=== Sample Data (First 5 Rows) ===")
print(df.head())
```

```python
#Remove duplicates if present.
import pandas as pd

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")
# Check duplicate rows BEFORE removal
print("Duplicate rows BEFORE removal:")
print(df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Check duplicate rows AFTER removal
print("\nDuplicate rows AFTER removal:")
print(df.duplicated().sum())

# Dataset shape after removing duplicates
print("\nDataset shape after removing duplicates:")
print(df.shape)

# Preview cleaned data
print("\nFirst 5 rows after duplicate removal:")
print(df.head())
```
