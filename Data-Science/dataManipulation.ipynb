```python
#1.	Filtering rows based on conditions (e.g., age > 30).
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")

# Filter rows where Age > 30 and sex is female
age = df[(df["Age"] > 30) & (df["Sex"] == "female")]

# Display results
print("Passengers with Age > 30:")
print(age.head())

print("\nNumber of passengers with Age > 30:")
print(age.shape[0])
```

```python
#Selecting columns & using .loc / .iloc.
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")

# Select columns
print("Age column:")
print(df["Age"].head())

print("\nAge and Fare columns:")
print(df[["Age", "Fare"]].head())

# Using .loc
print("\nUsing .loc (rows 0 to 4, Age & Fare):")
print(df.loc[0:4, ["Age", "Fare"]])

# Using .iloc
print("\nUsing .iloc (first 5 rows, first 2 columns):")
print(df.iloc[:5, :2])
```

```python
#Grouping & Aggregations
# Mean, count, sum, min/max of relevant columns.

import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")

# Group by Passenger Class and calculate aggregations
data = df.groupby("Pclass")["Fare"].agg(
    mean_fare="mean",
    total_fare="sum",
    min_fare="min",
    max_fare="max",
    count_passengers="count"
)

# Display result
print(data)
```

```python
#4.	Merging/Joining 
# (if dataset has multiple CSVs, or create a small second dataset and merge them).
import pandas as pd
df1 = pd.read_csv("Titanic-Dataset.csv")

# Create a small second dataset
df2 = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Name": ["First Class", "Second Class", "Third Class"]
})

# Merge datasets on Pclass
merged_df = pd.merge(df1, df2, on="Pclass", how="left")

# Display result
print(merged_df[["Name", "Pclass", "Class_Name"]].head())
```
