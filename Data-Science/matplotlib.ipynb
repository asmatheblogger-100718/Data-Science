```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Titanic-Dataset.csv")

# Line Plot Trend of Fare (first 50 passengers)
plt.figure()
plt.plot(df["Fare"].head(50), color="blue", label="Fare Trend")
plt.title("Fare Trend (First 50 Passengers)")
plt.xlabel("Passenger Index")
plt.ylabel("Fare")
plt.legend()
plt.grid(True)
plt.show()

#Bar Chart  Passenger Count by Class
classP = df["Pclass"].value_counts()

plt.figure()
plt.bar(classP.index, classP.values, color="green")
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.grid(axis="y")
plt.show()

#  Histogram  Age Distribution
plt.figure()
plt.hist(df["Age"].dropna(), bins=20, color="orange", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Scatter Plot  Age vs Fare
plt.figure()
plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6, label="Age vs Fare")
plt.title("Age vs Fare Relationship")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
plt.grid(True)
plt.show()

# Pie Chart  Survival Distribution (Bonus)
survival_counts = df["Survived"].value_counts()

plt.figure()
plt.pie(
    survival_counts,
    labels=["Did Not Survive", "Survived"],
    autopct="%1.1f%%",
    colors=["gray", "skyblue"]
)
plt.title("Survival Distribution")
plt.show()
```
